```python
# Effect of SD
# Read the excel file and put the loss values for each S.D. case in lists
train_loss_filename = '/home/parika/WorkingDir/Thesis/Documentation/ejecta/EffectOfSD/Training_Loss.xlsx'
val_loss_filename = '/home/parika/WorkingDir/Thesis/Documentation/ejecta/EffectOfSD/Validation_Loss.xlsx'

train_loss_workbook = xlrd.open_workbook(train_loss_filename)
val_loss_workbook = xlrd.open_workbook(val_loss_filename)

# Read the training loss values
first_sheet = train_loss_workbook.sheet_by_index(0)

labels = [first_sheet.cell_value(0,i) for i in range(1,first_sheet.ncols)]
epochs = [first_sheet.cell_value(i,0) for i in range(1,first_sheet.nrows)]

train_loss_values = {}
for col in range(1,first_sheet.ncols):
    label = labels[col-1]
    loss = [first_sheet.cell_value(i,col) for i in range(1,first_sheet.nrows)]
    train_loss_values[label] = loss

# Read the validation loss values
first_sheet =val_loss_workbook.sheet_by_index(0)

val_loss_values = {}
for col in range(1,first_sheet.ncols):
    label = labels[col-1]
    loss = [first_sheet.cell_value(i,col) for i in range(1,first_sheet.nrows)]
    val_loss_values[label] = loss
```

```python
# Plot the read loss values in a graph
colors = ['#FF0033', '#FF00EF', '#0022FF', '#0000FF', '#FFAB00', '#3A2B0B', '#12D2B9', '#AB47BA', '#838B13', '#0E6614', '#FF00BC', '#7F202A']

prop_cycle = plt.rcParams['axes.prop_cycle']

params = {'legend.fontsize': 50,
         'axes.labelsize': 80,
         'axes.titlesize': 80,
         'xtick.labelsize':60,
         'ytick.labelsize':60}
pylab.rcParams.update(params)


matplotlib.rcParams['mathtext.fontset'] = 'cm'
matplotlib.rcParams['mathtext.rm'] = 'serif'

plt.rcParams["font.family"] = "cmr10"

plt.figure(figsize=(80,16),linewidth=20)

# Plot training loss on left side
plt.subplot(1, 2, 1)

for idx, label in enumerate(labels):
    plt.plot(epochs[:150], train_loss_values[label][:150], label=label, color=colors[idx], linewidth=5)
    plt.grid(True, which='major', alpha=.3, linewidth=5)

# plt.xticks(fontsize=40)
# plt.yticks(fontsize=40)
# plt.xlabel('Number of Epochs',fontsize=50)
# plt.ylabel('Training Loss',fontsize=50)
plt.xlabel('Number of Epochs')
plt.ylabel('Training Loss')

# Plot validation loss on left side
plt.subplot(1, 2, 2)

for idx, label in enumerate(labels):
    plt.plot(epochs, val_loss_values[label], label=label, color=colors[idx], linewidth=5)
    plt.grid(True, which='major', alpha=.3, linewidth=5)

# plt.xticks(fontsize=40)
# plt.yticks(fontsize=40)
# plt.xlabel('Number of Epochs',fontsize=50)
# plt.ylabel('Validation Loss',fontsize=50)
plt.xlabel('Number of Epochs')
plt.ylabel('Validation Loss')


plt.legend(loc='center left',bbox_to_anchor=(1, 0.5),edgecolor='0.0')#,prop={'size':40})
# plt.tight_layout(pad=1.0)
plt.savefig('/home/parika/WorkingDir/Thesis/Documentation/ejecta/EffectOfSD/loss_curves.pdf', bbox_inches = 'tight')
plt.close()
# plt.show()
```
